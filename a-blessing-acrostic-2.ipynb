```python
blessing_one = open("a_blessing_james_wright.txt").read().split()
blessing_two = open("a_blessing_mary_oliver.txt").read().split()

initials_one = {}
initials_two = {}

for item in blessing_one:
    first_let_1 = item[0].lower()
    
    if first_let_1 not in initials_one:
        initials_one[first_let_1] = []
            
    initials_one[first_let_1].append(item)

for item in blessing_two:
    first_let_2 = item[0].lower()
    
    if first_let_2 not in initials_two:
        initials_one[first_let_1] = []

    initials_two[first_let_2].append(item)
    
import random
```

```python
for first_let_1 in "a blessing":
    word = random.choice(initials_one.get(first_let_1, [" "]))
    print(word.capitalize())
```

```python
for first_let_2 in "a blessing":
    word = random.choice(initials_two.get(first_let_2, [" "]))
    print(word.capitalize())
```

```python
for first_let_2 in "a blessing":
    word = random.choice(initials_two.get(first_let_2, first_let_1))
    print(word.capitalize())
```
